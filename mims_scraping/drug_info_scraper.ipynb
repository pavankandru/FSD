```python
import csv

with open('./mims_links.csv', 'r') as csvfile: 
        heading = next(csvfile)
        reader_obj = csv.reader(csvfile)
          
        for row in reader_obj:
            print(row[0], str("https://www.mims.com" + str(row[1])[:-13] + "type=full&mtype=generic"))
            
            driver.get("https://www.mims.com" + str(row[1])[:-13] + "type=full&mtype=generic")
            time.sleep(5)
            page_source = driver.page_source
            
            with open('./scraped_drug_info_html/'+ row[0] +'.txt', 'w') as file:
                file.write(page_source)
```

```python
import csv
import os

with open('./mims_links.csv', 'r') as csvfile: 
        heading = next(csvfile)
        reader_obj = csv.reader(csvfile)
          
        for row in reader_obj:
            if os.path.exists('./scraped_drug_info_html/'+ row[0] +'.txt'):
                continue
            row[0] = row[0].replace("/", "OR")
            print(row[0], str("https://www.mims.com" + str(row[1])[:-13] + "type=full&mtype=generic"))
            
            driver.get("https://www.mims.com" + str(row[1])[:-13] + "type=full&mtype=generic")
            time.sleep(5)
            page_source = driver.page_source
            
            with open('./scraped_drug_info_html/'+ row[0] +'.txt', 'w') as file:
                file.write(page_source)
```
